<h3>Naravni prirast</h3>

Za naravni prirast smo uporabili 2 zbirki podatkov. Eno s podatki od leta 1995 do 2020 in vsebuje podatke za občine in eno, s podatki od leta 1922 do 2020, ki vsebuje podatke za celotno Slovenijo.

Razlike med številom prebivalcev v občinah so lahko ogromne. Za primerjavo - Ljubljana leta 2020 - 294050 prebivalcev, Hodoš leta 2020 pa le 360. Zato smo kot demografski kazalnik uporabljali "Naravni prirast na 1000 prebivalcev". To je razmerje med naravnim prirastom v koledarskem letu in številom prebivalstva sredi istega leta na določenem območju pomnoženo s 1000. Oziroma:
<center>$np=\frac{NP}{P}*1000$</center>

**NP** - naravni prirast<br>
**P** - število prebivalstva
